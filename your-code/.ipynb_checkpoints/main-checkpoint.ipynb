#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
users_table = pd.read_csv('datasets/users_table.csv')
users_table
```

#### 3. Rename Id column to userId

```python
users_table = users_table.rename(columns={'Id':'userId'})
users_table
```

#### 4. Import posts table:

```python
posts_table = pd.read_csv('datasets/posts_table.csv')
posts_table
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts_table = posts_table.rename(columns={'Id':'postId','OwnerUserId':'userId'})
posts_table.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users_columns = users_table[['userId','Reputation','Views','UpVotes','DownVotes']]
users_columns
```

```python
posts_columns = posts_table[['postId', 'Score','userId','ViewCount','CommentCount']]
posts_columns
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merge_df = pd.merge (left = users_columns,
                     right = posts_columns,
                     how = 'inner',
                     left_on = 'userId',
                     right_on = 'userId')

merge_df    
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
merge_null = merge_df.isnull().sum()
display(merge_null)
list(merge_null[merge_null>0].index)
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
#table showing only the ViewCount null
view_null = merge_df['ViewCount'].isnull()
merge_df[view_null].sample(20)
```

```python
#table showing only the ViewCount that are not null
view_notnull = merge_df['ViewCount'].notnull()
merge_df[view_notnull].sample(20)
```

```python
#The missing data comes from merging the "posts_table" and the "users_table" files.
#The data that we already have could be useful, so we replace the NaN values and turn it into the mean, so we can continue to make calculations.
posts_columns.isnull().sum()
```

```python
merge_df['ViewCount'] = merge_df['ViewCount'].fillna(merge_df['ViewCount'].mean())
merge_df
```

#### 10. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
merge_df['Reputation']= merge_df['Reputation'].astype('int64')
merge_df['Views']= merge_df['Views'].astype('int64')
merge_df['UpVotes']= merge_df['UpVotes'].astype('int64')
merge_df['DownVotes']= merge_df['DownVotes'].astype('int64')
merge_df['postId']= merge_df['postId'].astype('int64')
merge_df['Score']= merge_df['Score'].astype('int64')
merge_df['ViewCount']= merge_df['ViewCount'].astype('int64')
merge_df['CommentCount']= merge_df['CommentCount'].astype('int64')
```
